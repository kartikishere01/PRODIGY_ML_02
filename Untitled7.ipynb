```python
import pandas as pd
import os

file_path = os.path.join(path, "Mall_Customers.csv")
df = pd.read_csv(file_path)

df.head()
```

```python
df.columns = ['CustomerID', 'Gender', 'Age', 'AnnualIncome', 'SpendingScore']
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['AnnualIncome'], bins=20)
plt.title("Annual Income Distribution")
plt.show()

plt.hist(df['SpendingScore'], bins=20)
plt.title("Spending Score Distribution")
plt.show()
```

```python
plt.scatter(df['AnnualIncome'], df['SpendingScore'])
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.show()
```

```python
X = df[['AnnualIncome', 'SpendingScore']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    sil_scores.append(silhouette_score(X_scaled, labels))

plt.plot(range(2,11), sil_scores, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Analysis")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
plt.scatter(df['AnnualIncome'], df['SpendingScore'], c=df['Cluster'])
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segments")
plt.show()
```

```python
df.groupby('Cluster')[['AnnualIncome','SpendingScore']].mean()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_scaled, df['Cluster'])
score
```

```python
model_score = (score + 1) / 2 * 100
model_score
```

```python
import joblib

joblib.dump(kmeans, "kmeans_model.pkl")
joblib.dump(scaler, "scaler.pkl")
```
